```python
import numpy as np
import random
import ColabTurtle.Turtle as bot
```

```python
pen_color1 = "blue"
pen_color2 = "red"
pen_color3 = "green"

# image ratio
ratio = 1

# left eye
# change last angle
left_eye_len = np.array([40, 40, 15, 35, 17, 15, 35]) * ratio
left_eye_deg = np.array([10, 60, 66, 71, 35, 112, 50])
left_eye_dir = np.array([False, True, True, True, False, False, False])
left_eye_forward = np.ones(len(left_eye_len), dtype=bool)

# right eye + nose
right_eye_len = np.array([25, 27, 30, 30, 54, 10, 24, 10]) * ratio
right_eye_deg = np.array([8, 17, 157, 38, 80, 63, 45, 54])
right_eye_dir = np.array([True, False, False, False, False, True, True, True])
right_eye_forward = np.ones(len(right_eye_len), dtype=bool)

# face
# change from 14th
face_len = np.array([15, 20, 17, 54, 27, 27, 40, 46, 37, 74, 68, 78, 80, 50,
                      70, 45, 57, 95, 107, 103, 90]) * ratio
face_deg = np.array([60, 16, 30, 147, 159, 130, 123, 55, 25, 25, 30, 44, 54,
                      48, 24, 104, 84, 16, 50, 59, 40])
face_dir = np.array([True, True, True, True, False, True, False, False, False,
                      False, False, False, False, False, False, True, True,
                      True, True, True, True])
face_forward = np.ones(len(face_len), dtype=bool)

# travel path when pen is up
pen_up_len = [46, 48] * ratio
pen_up_deg = [5, 40]
pen_up_dir = [True, True]
pen_up_forward = np.ones(len(pen_up_len), dtype=bool)
pen_up_forward[1] = False
```

```python
def draw(length, deg, dir, forward, color):
  bot.color(color)
  for i in range(len(length)):
    if (dir[i]):
      bot.right(float(deg[i]))
    else:
      bot.left(float(deg[i]))
    if (forward[i]):
      bot.forward(float(length[i]))
    else:
      bot.backward(float(length[i]))
    # print(bot.position())
```

```python
def randomL(length, ind = True):
  if (ind):
    newlength = []
    for i in length:
      if i > 30:
        newlength.append(round(i + random.randint(-5, 5)))
      else:
        newlength.append(i)
    return newlength
  else:
    return length

def randomA(angle, ind = True):
  if (ind):
    newAngle = []
    for i in angle:
      if i < 30:
        newAngle.append(i + random.randint(0, 5))
      elif i < 150:
        newAngle.append(i + random.randint(-3, 3))
      else:
        newAngle.append(i - random.randint(0, 5))
    return newAngle
  else:
    return angle
```

```python
# expected drawing
ratio = 0.7
bot.initializeTurtle(initial_speed=5)

# initial position
bot.penup()
bot.setx(320)
bot.sety(120)
bot.pendown()
print(bot.position())

# set initial position
bot.right(90)

draw(left_eye_len * ratio, left_eye_deg, left_eye_dir, left_eye_forward, pen_color1)

bot.penup()
bot.right(10) # changed from 5
bot.forward(46 * ratio)
bot.pendown()

draw(right_eye_len * ratio, right_eye_deg, right_eye_dir, right_eye_forward, pen_color2)

bot.penup()
bot.right(43)
bot.backward(30 * ratio)
bot.pendown()

draw(face_len * ratio, face_deg, face_dir, face_forward, pen_color3)

# final position
print(bot.position())
```

```python
# expected drawing with randomization
ratio = 1

bot.initializeTurtle(initial_speed=5)
indA = True
indL = False

# set initial position
bot.penup()
bot.setheading(0)
bot.setx(300)
bot.sety(120)
bot.pendown()

draw(randomL(left_eye_len, indL) * ratio, randomA(left_eye_deg, indA),
      left_eye_dir, left_eye_forward, pen_color1)

bot.penup()
bot.right(randomA([10], indL)[0]) # changed from 5
bot.forward(randomL([46], indA)[0] * ratio)
bot.pendown()

draw(randomL(right_eye_len, indL) * ratio, randomA(right_eye_deg, indA),
    right_eye_dir, right_eye_forward, pen_color2)

bot.penup()
bot.right(randomA([43], indA)[0])
bot.backward(randomL([30], indL)[0] * ratio)
bot.pendown()

draw(randomL(face_len, indL) * ratio, randomA(face_deg, indA),
    face_dir, face_forward, pen_color3)
```

```python
# expected drawing with multiple iterations

count = 10
bot.initializeTurtle(initial_speed=13)
indL = True # random length
indA = False # random angle

for i in range(count):
  # set initial position
  bot.setheading(0)
  bot.pendown()

  draw(randomL(left_eye_len, indL) * ratio, randomA(left_eye_deg, indA),
      left_eye_dir, left_eye_forward, pen_color1)

  bot.penup()
  bot.right(randomA([10], indL)[0]) # changed from 5
  bot.forward(randomL([46], indA)[0] * ratio)
  bot.pendown()

  draw(randomL(right_eye_len, indL) * ratio, randomA(right_eye_deg, indA),
      right_eye_dir, right_eye_forward, pen_color2)

  bot.penup()
  bot.right(randomA([43], indA)[0])
  bot.backward(randomL([30], indL)[0] * ratio)
  bot.pendown()

  draw(randomL(face_len, indL) * ratio, randomA(face_deg, indA),
      face_dir, face_forward, pen_color3)

  bot.penup()
  bot.setx(400)
  bot.sety(250)

  i = i + 1
```
